```python
import pandas as pd
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
URL = "https://www.bls.gov/news.release/archives/empsit_10022020.htm"
tables = pd.read_html(URL)

print(len(tables)," tables")
household_data = tables[0]
```

```python
status = household_data[1:9]
unemp_rates = household_data[11:25].drop([19])
reason_unemp = household_data[27:31]
duration = household_data[33:37]
part_time = household_data[39:47].drop([43,44])
non_labor = household_data[45:47]
```

```python
subtitle = "From 8/20 - 9/20"

ax1 = sns.barplot(y="Category", x="Change from:Aug.2020-Sept.2020", 
                 data=unemp_rates.sort_values("Change from:Aug.2020-Sept.2020",
                                              ascending=False))

ax1.set_title("Unemployment rates {}".format(subtitle))
```

```python
ax2 = sns.barplot(y="Category", x="Change from:Aug.2020-Sept.2020", 
                 data=duration.sort_values("Change from:Aug.2020-Sept.2020",
                                              ascending=False))

ax2.set_title("Unemployment Duration {}".format(subtitle))
```

```python
ax3 = sns.barplot(y="Category", x="Change from:Aug.2020-Sept.2020", 
                 data=non_labor.sort_values("Change from:Aug.2020-Sept.2020",
                                              ascending=False))

ax3.set_title("Not in Labor Force {}".format(subtitle))
```

```python
ax4 = sns.barplot(y="Category", x="Change from:Aug.2020-Sept.2020", 
                 data=part_time.sort_values("Change from:Aug.2020-Sept.2020",
                                              ascending=False))

ax4.set_title("Part Time {}".format(subtitle))
```
